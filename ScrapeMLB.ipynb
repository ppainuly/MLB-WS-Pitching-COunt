## Scrapping Script to get MLB Datasets. Two different datasets are gathered and stored in a dataframe, which then exports out as .csv files.

#### The two datasets scrapped : Pitching Stats Regular season 2019, Expanded Stats II Regular season 2019

#### Aim : To gather every single pitch faced by Houston Astros and Washington Nationals.

```python
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
import re
from pprint import pprint
import numpy as np
```

```python
# Part 1: Setting up url to scrape
# #Mac Users
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

# Windows Users
#executable_path = {'executable_path': 'chromedriver.exe'}
#browser = Browser('chrome', **executable_path, headless=False)
```

## Get Basic Pitching Data for each team(HOU and WAS)

```python
# Visit the browser
# Get Pitching Stats
# Executes scrapping for the specific team (based on the url passed)
# Returns a list or lists containing name,team,h and ip as separate items of the list
def count_pitching(url,team):
    
    ip = []
    name = []
    team = []
    h = []
    overall = []
    count = 1
    mainurl = url
    if team == 'hou':
        lim = 282
    else: 
        lim = 321
    
    
    while count <= lim:

        browser.visit(url)

        html = browser.html
        soup = BeautifulSoup(html, "html.parser")

        response= soup.find('tbody')
        
        tag = response.findAll('tr')[2:]


        for tr in tag:
            print("Scrapping Pitching data")
            print(tr.findAll('td')[0].text)
            print(tr.findAll('td')[1].text) # player name
            name.append(tr.findAll('td')[1].text)
            print(tr.findAll('td')[2].text) # team name
            team.append(tr.findAll('td')[2].text)
            print(tr.findAll('td')[5].text) # IP
            ip.append(tr.findAll('td')[5].text)
            print(tr.findAll('td')[6].text) # H
            h.append(tr.findAll('td')[6].text)
            print('')
        if (count == 281) and (team == 'hou'):
            count+=1
        else:
            count+=40
        url = mainurl[:-4] + "count/" + str(count) + "/qualified/false/order/false"
        print(url)
    overall.append(name)
    overall.append(team)
    overall.append(h)
    overall.append(ip)
    return overall
    
```

```python
# Get Second pitching data

def count_pitching2(url,team):
    mainurl1 = ''
    mainurl2 = ''
    pit = []
    pip = []
    name = []
    team = []
    overall = []
    count = 1
    if team == 'hou':
        lim = 282
        mainurl1 = 'http://www.espn.com/mlb/stats/pitching/_/seasontype/2/split/18/count/'
        mainurl2 = '/qualified/false/type/expanded-2/order/false'
    else: 
        lim = 321
        mainurl1 = 'http://www.espn.com/mlb/stats/pitching/_/seasontype/2/split/20/count/'
        mainurl2 = '/qualified/false/type/expanded-2/order/false'
    
    
    while count <= lim:

        browser.visit(url)

        html = browser.html
        soup = BeautifulSoup(html, "html.parser")

        response= soup.find('tbody')
        
        tag = response.findAll('tr')[2:]


        for tr in tag:
            print(tr.findAll('td')[1].text) # player name
            name.append(tr.findAll('td')[1].text)
            print(tr.findAll('td')[6].text)
            pit.append(tr.findAll('td')[6].text)
            print(tr.findAll('td')[2].text) # team name
            team.append(tr.findAll('td')[2].text)
            print(tr.findAll('td')[8].text)
            pip.append(tr.findAll('td')[8].text)
        if (count == 281) and (team == 'hou'):
            count+=1
        else:
            count+=40
        url = mainurl1 + str(count) + mainurl2
        print(url)
    overall.append(name)
    overall.append(team)
    overall.append(pit)
    overall.append(pip)
    return overall
```

```python
# Scrape Houston Data - All Pitching against HOU regular season
url = "http://www.espn.com/mlb/stats/pitching/_/seasontype/2/split/18/sort"
hou_pitching_1 = count_pitching(url, 'hou')
```
